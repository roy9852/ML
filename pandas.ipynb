<font color = pink> apply lambda 로 데이터 가공

```python
def get_sqaure(x):
    return x**2

print(f'3의 제곱은 {get_sqaure(3)} 입니다')
```

```python
lambda_square = lambda x : x**2
print(f'3의 제곱은 {lambda_square(3)} 입니다.')
```

```python
a = [1, 2, 3]
squares = list(map(lambda_square, a))
print(squares)
```

```python
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name', 'Name_len']].head(10)
```

```python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x <= 15 else 'Adult')
titanic_df[['Age', 'Child_Adult']].head(10)
```

```python
titanic_df['Child_Adult_Elderly'] = titanic_df['Age'].apply(lambda x : 'Child' if x <= 15 
                                                            else ('Adult' if x <= 60 else 'Elderly'))
titanic_df['Child_Adult_Elderly'].value_counts()
```

```python
def get_category(age):
    cat = ''
    if age <= 5: cat = 'Baby'
    elif age <= 12 : cat = 'Child'
    elif age <= 18 : cat = 'Teenager'
    elif age <= 25 : cat = 'Student'
    elif age <= 35 : cat = 'Young Adult'
    elif age <= 60 : cat = 'Adult'
    else : cat = 'Elderly'
    
    return cat

titanic_df['Age_category'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age', 'Age_category']].head(20)
```
